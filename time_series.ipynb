```python
import numpy as np # linear algebra
from numpy import newaxis
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from keras.layers import Dense, Activation, Dropout
from keras.layers import LSTM, GRU
from keras.models import Sequential
from keras import optimizers
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
Enrol_window = int(50)
```

```python
# Support functions
sc = MinMaxScaler(feature_range=(0,1))
def load_data(datasetname, column, seq_len, normalise_window):
    # A support function to help prepare datasets for an RNN/LSTM/GRU
    # data = datasetname.loc[:,column]
    data  = np.array(datasetname['Close'])
    sequence_length = seq_len + 1
    result = []
    for index in range(len(data) - sequence_length):
        result.append(data[index: index + sequence_length])

    # if normalise_window:
    #     #
    #     result = normalise_windows(result)

        #
        # df = scaler.fit_transform(df)

    result = np.array(result)
    scaler = MinMaxScaler(feature_range=(0, 1))
    result = scaler.fit_transform(result)
    print(result.shape)

    #Randomly shuffling values
    # np.random.shuffle(result)

    #After shuffling, last 10% is used for validation test, first 90% for training
    row = round(0.9 * result.shape[0])
    train = result[:int(row), :]
    np.random.shuffle(train)
    x_train = train[:, :-1]
    y_train = train[:, -1]
    x_test = result[int(row):, :-1]
    y_test = result[int(row):, -1]

    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

    return [x_train, y_train, x_test, y_test]

```

```python
def normalise_windows(window_data):
    # A support function to normalize a dataset
    normalised_data = []
    for window in window_data:
        normalised_window = [((float(p) / float(window[0])) - 1) for p in window]
        normalised_data.append(normalised_window)
    return normalised_data
```

```python
def plot_results(predicted_data, true_data):
    fig = plt.figure(figsize=(25,25),facecolor='white')
    ax = fig.add_subplot(111)
    ax.plot(true_data, label='True Data')
    plt.plot(predicted_data, label='Prediction')
    plt.legend()
    plt.show()
```

```python
feature_train, label_train, feature_test, label_test = load_data(data, 'Close', Enrol_window, True)
```

```python
# CNN stacked LSTM model
model = Sequential()
model.add(tf.keras.layers.Conv1D(16, kernel_size=3, activation='tanh', input_shape=(feature_train.shape[1],1)))
model.add(tf.keras.layers.MaxPooling1D(pool_size=2))
model.add(tf.keras.layers.Conv1D(32, kernel_size=3, activation='tanh'))
model.add(tf.keras.layers.MaxPooling1D(pool_size=2))
model.add(tf.keras.layers.Flatten())

# Add LSTM layer
model.add(tf.keras.layers.Reshape((1,-1)))
model.add(tf.keras.layers.LSTM(64))

# Add output layer
model.add(Dense(16,activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='relu'))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.005)
model.compile(loss='mse', optimizer=optimizer,metrics=[tfa.metrics.RSquare()])

model.summary()
```

```python
# Add ConvLSTM layers
model = Sequential()
model.add(tf.keras.layers.Reshape((1,feature_train.shape[1],1), input_shape=(feature_train.shape[1],1)))
model.add(tf.keras.layers.ConvLSTM1D(filters=64, kernel_size=3, activation='relu', return_sequences=True))
model.add(tf.keras.layers.BatchNormalization())
model.add(Dropout(0.1))
model.add(tf.keras.layers.ConvLSTM1D(filters=64, kernel_size=3, activation='relu', return_sequences=True))
model.add(tf.keras.layers.BatchNormalization())
model.add(Dropout(0.1))
model.add(tf.keras.layers.ConvLSTM1D(filters=64, kernel_size=3, activation='relu', return_sequences=False))
model.add(tf.keras.layers.BatchNormalization())
model.add(Dropout(0.1))
# Flatten the output of ConvLSTM
model.add(tf.keras.layers.Flatten())

# Add output layer
model.add(Dropout(0.2))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='relu'))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.005)
model.compile(loss='mse', optimizer=optimizer,metrics=[tfa.metrics.RSquare()])

model.summary()
```

```python
#Train the model
history = model.fit(feature_train, label_train, epochs=5, validation_data = (feature_test, label_test))
```

```python
def evaluate():
    # Input the csv file
    """
    Sample evaluation function
    Don't modify this function
    """
    df = pd.read_csv('/content/drive/MyDrive/EE798/Sample/sample_input.csv')
    # df = dftata

    actual_close = np.loadtxt('/content/drive/MyDrive/EE798/Sample/sample_close.txt')
    # actual_close = dfact


    pred_close = predict_func(df)

    # Calculation of squared_error
    actual_close = np.array(actual_close)
    pred_close = np.array(pred_close)
    mean_square_error = np.mean(np.square(actual_close-pred_close))


    pred_prev = [df['Close'].iloc[-1]]
    pred_prev.append(pred_close[0])
    pred_curr = pred_close

    actual_prev = [df['Close'].iloc[-1]]
    actual_prev.append(actual_close[0])
    actual_curr = actual_close

    # Calculation of directional_accuracy
    pred_dir = np.array(pred_curr)-np.array(pred_prev)
    actual_dir = np.array(actual_curr)-np.array(actual_prev)
    dir_accuracy = np.mean((pred_dir*actual_dir)>0)*100

    print(f'Mean Square Error: {mean_square_error:.6f}\nDirectional Accuracy: {dir_accuracy:.1f}')


```

```python
def predict_func(data):
    """
    Modify this function to predict closing prices for next 2 samples.
    Take care of null values in the sample_input.csv file which are listed as NAN in the dataframe passed to you
    Args:
        data (pandas Dataframe): contains the 50 continuous time series values for a stock index

    Returns:
        list (2 values): your prediction for closing price of next 2 samples
    """

    # Filling NaN Values using Average, Backward Fill and Forward Fill
    data['Close'].fillna(-9999, inplace=True)
    na_indices = data['Close'].index[data['Close'] == -9999]

    for i in na_indices:
      # If first and last value is na
      if i != 0 and i!=49:
        data['Close'][i] = data['Close'][i-1]/2 + data['Close'][i+1]/2

    df=data['Close']

    # Doing bfill and ffill if any value boundary value is left
    df.bfill()
    df.ffill()



    df=np.array(df)
    df=np.reshape(df,(50, 1))

    # Scaling our data in range 0 to 1 because
    # our model is trained on 0 to 1
    scaler = MinMaxScaler(feature_range=(0, 1))
    df = scaler.fit_transform(df)


    # Loading the model which is trained earlier
    model = tf.keras.models.load_model('ConvLSTM_3.h5')
    i=0

    #Predicting the value and appending it
    # to dataframe to predict next value
    while (i<=1):
      df=np.reshape(df,(1,50+i,1))
      prediction=model.predict(df[:,i:,:])
      prediction=np.array(prediction)
      df=np.reshape(df,(50+i,1))
      df=np.append(df, prediction)
      i=i+1

    df=np.reshape(df,(52,1))
    df=scaler.inverse_transform(df)

    return [df[-2][0], df[-1][0]]


```
